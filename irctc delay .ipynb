```python
import pandas as pd

# load dataset
df = pd.read_csv("train_delay_data.csv")

# check structure
df.info()
df.head()

# check missing values
df.isnull().sum()

# fill missing delays
df['Average_Delay(min)'] = df['Average_Delay(min)'].fillna(0)

# remove duplicates
df = df.drop_duplicates()

# convert to numeric
df['Average_Delay(min)'] = pd.to_numeric(df['Average_Delay(min)'])

# create delay category
df['Delay_Category'] = pd.cut(
    df['Average_Delay(min)'],
    bins=[0,15,60,999],
    labels=['On Time','Moderate Delay','Heavy Delay']
)

# save clean dataset
df.to_csv("clean_train_delay_data.csv", index=False)
```

# load dataset


```python
df = pd.read_csv("etrain_delays.csv")
```

```python
df.info()
df.head()
```

# check missing values



```python
df.isnull().sum()
```

# converting it to numeric

```python
df['average_delay_minutes'] = pd.to_numeric(
    df['average_delay_minutes'],
    errors='coerce'
)
```

```python
df['average_delay_minutes'] = df['average_delay_minutes'].fillna(
    df['average_delay_minutes'].median()
)
```

# remove duplicates


```python
df = df.drop_duplicates()
```

# creating delay category

```python
df['Delay_Category'] = pd.cut(
    df['average_delay_minutes'],
    bins=[0,15,60,999],
    labels=['On Time','Moderate Delay','Heavy Delay']
)
```

```python
df.to_csv("clean_train_delay_data.csv", index=False)
```

```python
# Convert minutes to hour-minute format
df['Delay_Time_Format'] = df['average_delay_minutes'].apply(
    lambda x: f"{int(x//60)}h {int(x%60)}m"
)
```
